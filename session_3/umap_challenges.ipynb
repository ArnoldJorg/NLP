```python
import gensim.downloader as api
import umap
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import for 3D plotting

# Load pre-trained word vectors
word_model = api.load("glove-wiki-gigaword-100")  # 100D word vectors

# Pick words to visualize
words = ["cold", "freezing", "space", "galaxy", "sun", "moon", "friend", "foe"]
word_vectors = np.array([word_model[word] for word in words])

# Reduce dimensions using UMAP (100D → 3D)
reducer = umap.UMAP(n_components=3, random_state=42)
word_embedding = reducer.fit_transform(word_vectors)

# Create a 3D scatter plot
fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(111, projection="3d")  # Set 3D plot

ax.scatter(word_embedding[:, 0], word_embedding[:, 1], word_embedding[:, 2], color="red", alpha=0.7)

# Add word labels
for i, word in enumerate(words):
    ax.text(word_embedding[i, 0], word_embedding[i, 1], word_embedding[i, 2], word, fontsize=12)

# Labels & title
ax.set_title("UMAP on Word Vectors (100D → 3D)")
ax.set_xlabel("UMAP Dimension 1")
ax.set_ylabel("UMAP Dimension 2")
ax.set_zlabel("UMAP Dimension 3")  

plt.show()
```
